## 2. Features Generation

### 2.a. Period of Day

```python
from datetime import datetime

def get_period_day(date):
    """
    Calculate the period such as 'maÃ±ana', 'tarde' and 'noche', depending on datetime provided in date variable (date).

    Args:
        date (str): It is a string that represents the date and time.

    Returns:
        (str): The period day based on Date-I:
                - morning (between 5:00 and 11:59)
                - afternoon (between 12:00 and 18:59)
                - night (between 19:00 and 4:59)
    """
    date_time = datetime.strptime(date, '%Y-%m-%d %H:%M:%S').time()
    morning_min = datetime.strptime("05:00", '%H:%M').time()
    morning_max = datetime.strptime("11:59", '%H:%M').time()
    afternoon_min = datetime.strptime("12:00", '%H:%M').time()
    afternoon_max = datetime.strptime("18:59", '%H:%M').time()
    evening_min = datetime.strptime("19:00", '%H:%M').time()
    evening_max = datetime.strptime("23:59", '%H:%M').time()
    night_min = datetime.strptime("00:00", '%H:%M').time()
    night_max = datetime.strptime("4:59", '%H:%M').time()
    
    if(date_time >= morning_min and date_time <= morning_max):
        return 'maÃ±ana'
    elif(date_time >= afternoon_min and date_time <= afternoon_max):
        return 'tarde'
    elif(
        ((date_time >= evening_min) and (date_time <= evening_max)) or
        ((date_time >= night_min) and (date_time <= night_max))
        ):
        return 'noche'
```

```python
data['period_day'] = data['Fecha-I'].apply(get_period_day)
```

### 2.b. High Season

```python
from datetime import datetime

def is_high_season(fecha):
    """
    Based on the date of the flight, it defines whether it is a high season.

    Args:
        fecha (str): It is a string that represents the date and time of a flight.

    Returns:
        (Int): 1: If the date is during a high Season
               0: Otherwhise
    """
    fecha_dt = datetime.strptime(fecha, '%Y-%m-%d %H:%M:%S') 
    mes_dia = (fecha_dt.month, fecha_dt.day) 

    # Definir rangos de temporada alta sin importar el aÃ±o
    temporadas_1 = [
        ((12, 15), (12, 31)),  # 15-Dic al 31-Dic
        ((1, 1), (3, 3)),      # 1-Ene al 3-Mar
        ((7, 15), (7, 31)),    # 15-Jul al 31-Jul
        ((9, 11), (9, 30))     # 11-Sep al 30-Sep
    ]

    # Verificar si la fecha estÃ¡ dentro de alguno de los rangos
    for inicio, fin in temporadas_1:
        if inicio <= mes_dia <= fin:
            return 1
    
    return 0  # Si no estÃ¡ en ninguna temporada alta
```

```python
data['high_season'] = data['Fecha-I'].apply(is_high_season)
```

### 2.c. Difference in Minutes

```python
def get_min_diff(data):
    """
    Calculate difference in minutes between Data-O (Datetime of flight operation) 
    and Date-I (Scheduled datetime of the flight).

    Args:
        data (Dataframe): It is a Dataframe with both columns (Data-O and Data-I) 
                                 to calculate the difference.

    Returns:
        (Dataframe): The Dataframe with a new column that contains the difference in minutes
                     between columns Data-O and Date-I.
    """
    fecha_o = datetime.strptime(data['Fecha-O'], '%Y-%m-%d %H:%M:%S')
    fecha_i = datetime.strptime(data['Fecha-I'], '%Y-%m-%d %H:%M:%S')
    min_diff = ((fecha_o - fecha_i).total_seconds())/60
    return min_diff
```

```python
data['min_diff'] = data.apply(get_min_diff, axis = 1)
```

### 2.d. Delay

```python
threshold_in_minutes = 15
data['delay'] = np.where(data['min_diff'] > threshold_in_minutes, 1, 0)
```

## 3. Data Analysis: Second Sight

### How is the delay rate across columns?

```python
def get_rate_from_column(data, column):
    delays = {}
    for _, row in data.iterrows():
        if row['delay'] == 1:
            if row[column] not in delays:
                delays[row[column]] = 1
            else:
                delays[row[column]] += 1
    total = data[column].value_counts().to_dict()
    
    rates = {}
    for name, total in total.items():
        if name in delays:
            rates[name] = round(total / delays[name], 2)
        else:
            rates[name] = 0
            
    return pd.DataFrame.from_dict(data = rates, orient = 'index', columns = ['Tasa (%)'])
```

```python
destination_rate = get_rate_from_column(data, 'SIGLADES')
destination_rate = destination_rate.sort_values(by='Tasa (%)', ascending=False)
destination_rate_values = data['SIGLADES'].value_counts().index
plt.figure(figsize = (20,5))
palette = sns.color_palette("viridis", len(destination_rate_values))
sns.barplot(x=destination_rate_values, y=destination_rate['Tasa (%)'], palette=palette, alpha = 1)
plt.title('Delay Rate by Destination')
plt.ylabel('Delay Rate [%]', fontsize=12)
plt.xlabel('Destination', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
airlines_rate = get_rate_from_column(data, 'OPERA')
airlines_rate = airlines_rate.sort_values(by='Tasa (%)', ascending=False)
airlines_rate_values = data['OPERA'].value_counts().index
plt.figure(figsize = (20,5))
palette = sns.color_palette("viridis", len(destination_rate_values))
sns.barplot(x=airlines_rate_values, y=airlines_rate['Tasa (%)'],palette=palette, alpha = 1)
plt.title('Delay Rate by Airline')
plt.ylabel('Delay Rate [%]', fontsize=12)
plt.xlabel('Airline', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```
